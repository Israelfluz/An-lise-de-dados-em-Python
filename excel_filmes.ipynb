# Análise de Dados de filmes exibidos no Brasil de 2009 a 2017
Com base nos dados da ancine http://oca.ancine.gov.br vamos analisar os filmes exibidos no Brasil

Arquivo -> 2120_0.xlsx

```python
# Importando as bibliotecas 
import pandas as pd

# Criando graficos com matplotlib
%matplotlib inline
```

```python
# Criando o DataFrame com base em um arquivo Excel
# Pulando as 2 primeiras linha do DataFrame que não são necessárias com (Skiprows)
df = pd.read_excel('2120_0.xlsx', skiprows=2, skipfooter=16)

# Exibindo as 5 primeiras linhas
df.head()
```

```python
# Exibindo as 5 últimas linhas
df.tail(20)
```

```python
# Verificando todos os  tipos de colunas
df.dtypes
```

```python
# Convertendo um dados específico (convertendo os Dados de Data de lançamento em data mesmo)
df['Data de lançamento'] = pd.to_datetime(df['Data de lançamento'])
```

```python
# Identificando o problema na coluna dar de lançamento
df['Data de lançamento'].value_counts()
```

```python
# Exibir alguns filmes que estão com data = relançamento 
df[df['Data de lançamento'] == 'Relançamento'].head()
```

## Decisão sobre dados incorretos/faltantes/divergentes

```python
# Verificando o tamanho do DataFrame (linhas, colunas)
df.shape
```

```python
# Criando um novo DataFrame sem os relançamentos
df_novo = df[df['Data de lançamento'] != 'Relançamento'].copy()
```

```python
df_novo.dtypes
```

```python
# Convertendo um dados específico (convertendo os Dados de Data de lançamento em data mesmo usando o to_datetime)
df_novo['Data de lançamento'] = pd.to_datetime(df_novo['Data de lançamento'])
```

```python
df_novo.dtypes
```

```python
# Quais os anos tiveram mais lançamentos?
df_novo['Data de lançamento'].dt.year.value_counts()
```

```python
# Imprimindo os dados acima em um gráfico
df_novo['Data de lançamento'].dt.year.value_counts().plot.barh()
```

```python
# Qual o filme teve a maior bilheteria
df_novo[df_novo['Renda (R$) no ano de exibição'] == df_novo['Renda (R$) no ano de exibição'].max()]
```

```python
# Os 3 filmes com maior bilheteria
df_novo.nlargest(3, 'Renda (R$) no ano de exibição')
```

```python
# Qual o filme teve a menor bilheteria
df_novo[df_novo['Renda (R$) no ano de exibição'] == df_novo['Renda (R$) no ano de exibição'].min()]
```

```python
# Quantos filmes brasileiros e estrangeiros/
df_novo['Nacionalidade da obra'].value_counts()
```
